```python
import pandas as pd
import matplotlib.pyplot as plt

# Load data (skip first 4 rows — World Bank files have a header block)
df = pd.read_csv("API_SP.POP.TOTL_DS2_en_csv_v2_38144.csv", skiprows=4)

print(df.shape)
print(df.columns.tolist())
df.head()
```

```python
# Pick the latest year with data (check last few columns for non-null values)
year = "2024"
pop_data = df[["Country Name", year]].dropna()
pop_data = pop_data[pop_data["Country Name"] != "World"]

```

```python
# Histogram - distribution of population sizes across countries
plt.hist(pop_data[year], bins=30, edgecolor='black')
plt.title(f"Distribution of Population Sizes Across Countries in {year}")
plt.xlabel("Population Size")
plt.ylabel("Number of Countries")
plt.show()
```

```python
# Bar chart - top 10 countries by population
top10 = pop_data.sort_values(by=year, ascending=False).head(10)
plt.figure(figsize=(10, 6))
plt.bar(top10["Country Name"], top10[year])
plt.xticks(rotation=60, ha='right')
plt.title(f"Top 10 Countries by Population in {year}")
plt.xlabel("Country")
plt.ylabel("Population Size")
plt.tight_layout()
plt.show()
```
